```python
Q.5. State advantages and disadvantages of multithreading
ANS:------>
Multithreading is a powerful feature that allows multiple threads to execute concurrently within a single program. Here are some advantages and disadvantages of multithreading:

Advantages:

1.Improved performance: By dividing a program into multiple threads, 
each thread can execute a different task simultaneously, which can lead to
faster execution times and improved performance.

2.Responsiveness: Multithreading allows a program to remain responsive even when 
one thread is blocked or performing a time-consuming task, as other threads can continue
executing in the background.

3.Resource sharing: Multithreading allows multiple threads to share resources such as memory,
CPU time, and I/O devices, which can lead to more efficient resource utilization.

4.Simplified programming: In some cases, multithreading can simplify programming by dividing complex
tasks into smaller, more manageable pieces that can be executed concurrently.

Disadvantages:

1.Increased complexity: Multithreading adds complexity to a program and can make it harder to debug
and maintain. Synchronization and communication between threads can also be challenging to implement correctly.

2.Resource contention: When multiple threads are accessing shared resources, contention for those
resources can lead to performance issues and may require additional synchronization
mechanisms to be put in place.

3.Deadlocks and race conditions: When multiple threads access shared resources
without proper synchronization, it can lead to deadlocks and race conditions that
can cause unpredictable behavior in a program.

4.Overhead: Multithreading requires additional overhead to manage the threads, including
thread creation and management, synchronization, and communication between threads. 
This can result in increased memory usage and reduced performance in some cases.

Overall, multithreading can be a powerful tool for improving performance and responsiveness 
in a program, but it also comes with added complexity and potential pitfalls that must be
carefully managed to ensure the correct functioning of a program.

Q.6. Explain deadlocks and race conditions.
ANS:------>
Deadlocks and race conditions are common issues that can occur in multithreaded programs.
Here's a brief explanation of each:

1.Deadlocks: A deadlock occurs when two or more threads are blocked, each waiting for
the other to release a shared resource. This can happen when two threads each hold a 
resource and try to acquire the other's resource before releasing its own, causing both
threads to become blocked indefinitely. Deadlocks can cause a program to hang or become
unresponsive, and can be difficult to detect and resolve.

2.Race conditions: A race condition occurs when two or more threads access a shared 
resource simultaneously, and the final outcome depends on the order in which the threads execute.
This can result in unpredictable behavior, as the output of the program may depend on the timing
of thread execution. Race conditions can be difficult to reproduce and debug, and can lead to 
subtle bugs that are hard to detect.

To avoid deadlocks and race conditions, it's important to use proper synchronization mechanisms
when accessing shared resources in a multithreaded program. This can include using locks or 
semaphores to ensure that only one thread can access a shared resource at a time, or using 
message passing to communicate between threads. It's also important to design the program
carefully to avoid situations where multiple threads could become blocked waiting for each other, 
or where the order of thread execution could affect the outcome of the program.

```
